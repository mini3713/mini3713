```python
from tslearn.clustering import TimeSeriesKMeans
```

### extract occupancy data

```python
for i in range(0, df_c5.shape[0]) : 
    
    for k in range(0, df_c5.shape[1]) : 
        
        a = str(df_c5.iloc[int(i), int(k)])
        
        if a[4:6] == '01' :
            
            df_c5.iloc[int(i), int(k)] = 1
            
        else : df_c5.iloc[int(i), int(k)] = 0
            
df_c5.head()
```

```python
df_c5.info()
temp = df_c5
```

### Clustering 

```python
km = TimeSeriesKMeans(n_clusters=4, metric="euclidean", random_state=42, n_jobs=-1)
y_pred = km.fit_predict(df_c5)
```

```python
km.inertia_
```

```python
import matplotlib.pyplot as plt
inertia = []
for k in range(2, 20):
  km = TimeSeriesKMeans(n_clusters=k, random_state=42, metric="euclidean", max_iter=2)
  km.fit(df_c5)
  inertia.append(km.inertia_)
plt.plot(range(2, 20), inertia)
plt.xlabel('k')
plt.ylabel('inertia')
plt.show()
```

```python
import matplotlib.pyplot as plt
inertia = []
for k in range(1, 10):
  km = TimeSeriesKMeans(n_clusters=7, random_state=42, metric="euclidean", max_iter=k)
  km.fit(df_c5)
  inertia.append(km.inertia_)
plt.plot(range(1, 10), inertia)
plt.xlabel('k')
plt.ylabel('inertia')
plt.show()
```

```python
# cluster labels 
tkm = TimeSeriesKMeans(n_clusters=7, random_state=42, metric="euclidean", max_iter=2)
cluster5 = tkm.fit(df_c5)
df_c5['cluster'] = cluster5.labels_
df_c5.head()
```
