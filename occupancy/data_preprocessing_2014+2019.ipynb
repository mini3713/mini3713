```python
import pandas as pd
```

```python
feature = pd.read_csv("/content/2014_시간대_전체_20230112_71326.csv", encoding="cp949")
feature2 = pd.read_csv("/content/2014_시간대_전체_20230112_37950.csv", encoding="cp949")
```

```python
# 2014 데이터를 2019 데이터 순서와 일치하도록 수정 
feature = feature[['행정구역시도코드','가구일련번호','가구원번호','조사요일코드','요일구분코드','만연령','연령코드','경제활동상태코드']]
work = feature2['근무등교코드']
number = feature2['조사차수']
```

```python
new_feat = pd.concat([number, feature, work], axis=1)
new_feat.head()
```

* 조사요일코드, 근무등교코드 다름 

```python
# 코드 일치 작업
for i in new_feat.index:
  val = new_feat.loc[i, '조사요일코드']
  if val > 0 : new_feat.loc[i, '조사요일코드'] = -1 + val 
  else : pass  
```

```python
for i in new_feat.index:
  val = new_feat.loc[i, '근무등교코드']
  if val == 3 : new_feat.loc[i, '근무등교코드'] = 2 
  else : pass  
```

```python
new_feat['근무등교코드'].value_counts()
```

```python
feat_2014 = new_feat[['조사차수','행정구역시도코드','조사요일코드','요일구분코드','연령코드','근무등교코드']]
```

```python
ff = pd.read_csv("/content/2014_시간대_전체_20230112_71326.csv", encoding="cp949")
new_ts = ff.iloc[:,8:]
new_ts
```

```python
# 2014 시계열 데이터를 2019 코드와 일치하는 작업
for i in range(0, new_ts.shape[0]) :     
    for k in range(0, new_ts.shape[1]) :         
        a = new_ts.iloc[int(i), int(k)]
        if a == 0 : new_ts.iloc[int(i), int(k)] = 1       
        else : new_ts.iloc[int(i), int(k)] = 0
            
new_ts.head()
```

```python
data_2014 =  pd.concat([feat_2014, new_ts], axis=1)
data_2014.to_excel(excel_writer='Time use survey_Korea_2014.xlsx')
```
